## Combine the Wolves Counted, Packs Monitored, and Area Surveyed into Single Columns in Order to Perform Plots to Answer Analysis Questions

```python
WolfCount=df.WolfCountSpring+df.WolfCountFall
```

```python
PackCount=df.PacksSpring+df.PacksFall
```

```python
Area=df.AreaSqrKMSpring+df.AreaSqrKMFall
```

## Add These New Columns to the Original Data Frame

```python
df['WolfCount']=WolfCount
```

```python
df['PackCount']=PackCount
```

```python
df['Area']=Area
```

## Create a New Data Frame With the Renamed Columns and New Columns

```python
df2=df[['Year','WolfCount','PackCount','Area','PacksSpring', 'PacksFall', 'WolfCountSpring', 'WolfCountFall','AreaSqrKMSpring','AreaSqrKMFall']]
```

## Save Data Frame of Cleaned Data to a New CSV File

```python
df2.to_csv('WolfClean.csv', index=False)
```
